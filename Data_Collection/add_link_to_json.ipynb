```python
import json
json_file_path = 'courses.json'
txt_file_path = 'filtered_file.txt'
output_file_path = 'combined_data.json'
```

```python
url_list = []

# Open the text file and read each line
with open(txt_file_path, 'r') as txt_file:
    for line in txt_file:
        # Strip any leading/trailing whitespace and newline characters
        url = line.strip()
        # Add the URL to the list if it's not empty
        if url:
            url_list.append(url)

# Print the list of URLs
url_list
```

```python
with open(json_file_path, 'r') as json_file:
    json_data = json.load(json_file)
    
```

```python
for i in range(len(json_data)):
    json_data[i]['Link'] = url_list[i]
```

```python
# Output the modified data
for course in json_data:
    print(course)
```

```python
# Save as a JSON file
with open('courses.json', 'w') as json_file:
    json.dump(json_data, json_file, indent=4)

print("Data saved to 'courses.json'.")
```
