```python
link = "http://nku.catalog.acalog.com/content.php?catoid=27&navoid=4058" 
```

```python
import pandas as  pd
import numpy as np
import requests as req
from bs4 import BeautifulSoup
```

```python
response = req.get(link)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
list_items = soup.find_all('li')
list_items
```

```python
results = []
```

```python
for li in list_items:
    a_tag = li.find('a')
    if a_tag:  # Check if <a> exists inside <li>
        results.append({
            'li_text': li.text.strip(),    # Get the text inside <li>
            'a_href': a_tag['href'],       # Get the href attribute of <a>
            'a_text': a_tag.text.strip()   # Get the text inside <a>
        })
```

```python
import re
def clean_text(text):
    # Remove bullet points, newlines, and leading/trailing whitespace
    cleaned = re.sub(r'•\s*|\n', '', text).strip()
    return cleaned
```

```python
for program in results:
    program['li_text'] = clean_text(program['li_text'])
```
