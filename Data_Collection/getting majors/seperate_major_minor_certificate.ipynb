```python
import pandas as pd
import numpy as np
import json
```

```python
filename = 'all_major_links.json'
```

```python
try:
    with open(filename, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)
    print(data)  # Print or process the loaded data
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
except FileNotFoundError:
    print(f"The file {filename} does not exist.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
data
```

```python
for program in data:
    if 'Certificate' in program['li_text']:
        program['type'] = 'certificate'
    elif 'Minor' in program['li_text']:
        program['type'] = 'minor'
    else:
        program['type'] = 'major'
```
