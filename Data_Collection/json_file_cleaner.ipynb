```python
import json
```

```python
with open('courses_sample.json', 'r') as file:
    data = json.load(file)
```

```python
data
```

```python
def convert_prereqs_and_coreqs(courses):
    for course in courses:
        # Convert Prerequisites
        if course['Prerequisites'] == '[]':
            course['Prerequisites'] = []
        else:
            # If there are actual values, convert the string to a list
            try:
                course['Prerequisites'] = eval(course['Prerequisites'])  # Use with caution
            except Exception:
                course['Prerequisites'] = []

        # Convert Coreq
        if course['Coreq'] == '[]':
            course['Coreq'] = []
        else:
            try:
                course['Coreq'] = eval(course['Coreq'])  # Use with caution
            except Exception:
                course['Coreq'] = []

    return courses
```

```python
converted_data = convert_prereqs_and_coreqs(data)
```
