```python
def get_links(response):
    html_content= response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    tables = soup.find_all('table', class_ = "table_default")
    last_table = tables[-1]  # Get the last table
    rows = last_table.find_all('tr')
    with open('course_urls.txt', 'a') as file:   
        for row in rows:
            # Find all 'td' tags in the row
            cells = row.find_all('td')
            for cell in cells:
                a_tag = cell.find('a')
                if a_tag and a_tag.has_attr('href'):
                    href = a_tag['href']
                    full_url = f"http://nku.catalog.acalog.com/{href}"
                    file.write(full_url + '\n')
```

```python
base_url = "http://nku.catalog.acalog.com/content.php?filter%5B27%5D=-1&filter%5B29%5D=&filter%5Bkeyword%5D=&filter%5B32%5D=1&cur_cat_oid=23&expand=&navoid=2904&search_database=Filter&filter%5Bexact_match%5D=1&filter%5Bcpage%5D="
total_pages = 28
```

```python
for page_num in range(1, total_pages + 1):
    full_url = f"{base_url}{page_num}#acalog_template_course_filter"
    response = req.get(full_url)
    get_links(response)
    print("Completed cycle for", total_pages)
```

```python
def remove_nth_urls(n, input_file='course_urls.txt', output_file='filtered_course_urls.txt'):
    with open(input_file, 'r') as file:
        urls = file.readlines()  # Read all the URLs into a list

    # Remove every nth URL (like 101, 202, 303, etc.)
    filtered_urls = [url for i, url in enumerate(urls, start=1) if i % n != 0]
    if filtered_urls:
        filtered_urls.pop()
    # Write the filtered URLs to a new file
    with open(output_file, 'w') as file:
        file.writelines(filtered_urls)
    
    print(f"Filtered URLs saved to {output_file}")

# Call the function to remove every 101st, 202nd, 303rd, etc.
remove_nth_urls(101)
```
